# Coeficiente de Pearson

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fang = pd.read_csv("../ficheiros/fang_data.csv")
fang.head()
```

```python
fang_cotacoes = fang[['Facebook', 'Amazon', 'Netflix', 'Google']]
```

```python
matriz_das_correlacoes = fang_cotacoes.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_das_correlacoes, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz das correlaes - Coeficientes de Pearson')
plt.show()
```

## Concluso
Existe uma correlao positiva entre os valores das aes do Facebook e da Amazon. e existe uma
correlao negativa entre a Amazon e a Netflix.
